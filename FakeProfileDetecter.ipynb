```python
temp = df[df['profile_pic']==False]

temp
```

```python
temp=df[df['private']=='Yes']
fake_count=temp['fake'].sum()
non_fake_count=temp.shape[0]-fake_count
counts= [fake_count, non_fake_count]
labels = ['Fake', 'Not-Fake']
plt.pie(counts, labels=labels, autopct="%.1f%%")
plt.axis('equal')
plt.title('Distribution of Fake vs. Not-Fake if the account is private')
```

```python
# Profile Picture Of Fake Vs Non Fake
fake_count = temp['fake'].sum()
non_fake_count = temp.shape[0] - fake_count
counts = [fake_count, non_fake_count]
labels = ['Fake', 'Not-Fake']
plt.pie(counts, labels=labels, autopct="%.1f%%")
plt.axis('equal')
plt.title('Fake vs. Not-Fake if they have a Profile Picture')
```

```python
#Added External URL 
temp=df[df["extern_url"]=='No']
fake_count = temp['fake'].sum()
non_fake_count = temp.shape[0] - fake_count
counts = [fake_count, non_fake_count]
label = ['Fake', 'Not-Fake']
plt.pie(counts, labels=label, autopct="%.1f%%")
plt.axis('equal')  
plt.title('Fake vs. Not-Fake if they have an External URL')
```
